```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
cancer =load_breast_cancer()
x_train,x_test,y_train,y_test = train_test_split(cancer.data,cancer.target, random_state=42)
mlp = MLPClassifier(random_state=0)
mlp.fit(x_train,y_train)
```

```python
print('Accuracy on the training subset is: {:.3f}'.format(mlp.score(x_train,y_train)))
print('Accuracy on the test subset is: {:.3f}'.format(mlp.score(x_test,y_test)))
```

```python
print('Maximum per each feature:\n{}'.format(cancer.data.max(axis=0)))
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit(x_train).transform(x_train)
x_test_scaled = scaler.fit(x_test).transform(x_test)
```

```python
mlp1 = MLPClassifier(max_iter=1000,random_state=0)
mlp1.fit(x_train_scaled,y_train)
print('Accuracy on the training subset is: {:.3f}'.format(mlp1.score(x_train_scaled,y_train)))
print('Accuracy on the test subset is: {:.3f}'.format(mlp1.score(x_test_scaled,y_test)))

```
