```python
from sklearn.svm import SVC
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
cancer = load_breast_cancer()
x_train,x_test,y_train,y_test = train_test_split(cancer.data,cancer.target,random_state=0)
svm = SVC()
svm.fit(x_train,y_train)
```

```python
print("Accuracy on training subsets is : {:.3f}".format(svm.score(x_train,y_train)))
print("Accuracy on test subsets is : {:.3f}".format(svm.score(x_test,y_test)))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(x_train.min(axis=0),'o',label='min')
plt.plot(x_train.max(axis=0),'v',label='max')
plt.xlabel('Feature Index')
plt.ylabel("Feature Magnetiude in Log Scale")
plt.yscale('log')
plt.legend(loc='upper right')
```

```python
min_train = x_train.min(axis=0)
range_train = (x_train - min_train).max(axis=0)
x_train_scaled = (x_train-min_train)/range_train
print('Minimum per feature\n : {}'.format(x_train_scaled.min(axis=0)))
print('Maximum per feature\n : {}'.format(x_train_scaled.max(axis=0)))
```

```python
x_test_scaled = (x_test - min_train)/range_train
```

```python
svm1 = SVC()
svm1.fit(x_train_scaled,y_train)
print("Accuracy on training subsets is : {:.3f}".format(svm1.score(x_train_scaled,y_train)))
print("Accuracy on test subsets is : {:.3f}".format(svm1.score(x_test_scaled,y_test)))
```

```python
svm2 = SVC(C =1000)
svm2.fit(x_train_scaled,y_train)
print("Accuracy on training subsets is : {:.3f}".format(svm2.score(x_train_scaled,y_train)))
print("Accuracy on test subsets is : {:.3f}".format(svm2.score(x_test_scaled,y_test)))
```
